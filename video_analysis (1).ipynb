```python
"""
Advanced Deepfake Video Dataset Analysis System
FaceForensics++ Dataset - Comprehensive Research Analysis
Author: Research Analysis System
Date: 2025-11-11
"""

# ============================================================================
# SECTION 1: AUTOMATIC DEPENDENCY MANAGEMENT & INSTALLATION
# ============================================================================

import sys
import subprocess
import importlib
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

def install_package(package_name, import_name=None):
    """Install package with error handling and verification"""
    if import_name is None:
        import_name = package_name
    
    try:
        importlib.import_module(import_name)
        print(f"‚úì {package_name} already installed")
        return True
    except ImportError:
        print(f"‚öô Installing {package_name}...")
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package_name])
            importlib.import_module(import_name)
            print(f"‚úì {package_name} successfully installed")
            return True
        except Exception as e:
            print(f"‚ö† Warning: Could not install {package_name}: {str(e)}")
            return False

# Install all required packages
print("=" * 80)
print("INITIALIZING DEEPFAKE ANALYSIS SYSTEM")
print("=" * 80)

required_packages = [
    ("numpy", "numpy"),
    ("pandas", "pandas"),
    ("matplotlib", "matplotlib"),
    ("seaborn", "seaborn"),
    ("opencv-python", "cv2"),
    ("scikit-learn", "sklearn"),
    ("scikit-image", "skimage"),
    ("scipy", "scipy"),
    ("pillow", "PIL"),
    ("tqdm", "tqdm"),
]

print("\n[1/7] Installing Core Dependencies...")
for package, import_name in required_packages:
    install_package(package, import_name)

# ============================================================================
# SECTION 2: IMPORT ALL LIBRARIES WITH ERROR HANDLING
# ============================================================================

print("\n[2/7] Importing Libraries...")

try:
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    import cv2
    from pathlib import Path
    from datetime import datetime
    import json
    import os
    from collections import defaultdict, Counter
    from typing import Dict, List, Tuple, Any
    import warnings
    from tqdm.notebook import tqdm
    
    # Scientific computing
    from scipy import stats
    from scipy.signal import welch
    from scipy.fft import fft2, fftshift
    
    # Image processing
    from skimage import exposure, filters, feature, measure
    from skimage.metrics import structural_similarity as ssim
    from PIL import Image
    
    # Machine learning
    from sklearn.decomposition import PCA
    from sklearn.manifold import TSNE
    from sklearn.preprocessing import StandardScaler
    from sklearn.ensemble import IsolationForest
    from sklearn.metrics import confusion_matrix, classification_report
    
    print("‚úì All libraries imported successfully")
    
except Exception as e:
    print(f"‚ö† Import error: {str(e)}")
    print("Attempting alternative imports...")

# ============================================================================
# SECTION 3: CONFIGURATION & SETUP
# ============================================================================

print("\n[3/7] Configuring Analysis Parameters...")

# Set plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 10
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12

# Create output directory for results
OUTPUT_DIR = Path("deepfake_analysis_results")
OUTPUT_DIR.mkdir(exist_ok=True)
print(f"‚úì Output directory created: {OUTPUT_DIR}")

# Dataset paths configuration
DATASET_PATHS = {
    'Original': r"C:\Users\chakr\Downloads\archive (3)\FaceForensics++_C23\original",
    'NeuralTextures': r"C:\Users\chakr\Downloads\archive (3)\FaceForensics++_C23\NeuralTextures",
    'FaceSwap': r"C:\Users\chakr\Downloads\archive (3)\FaceForensics++_C23\FaceSwap",
    'FaceShifter': r"C:\Users\chakr\Downloads\archive (3)\FaceForensics++_C23\FaceShifter",
    'Face2Face': r"C:\Users\chakr\Downloads\archive (3)\FaceForensics++_C23\Face2Face",
    'Deepfakes': r"C:\Users\chakr\Downloads\archive (3)\FaceForensics++_C23\Deepfakes",
    'DeepFakeDetection': r"C:\Users\chakr\Downloads\archive (3)\FaceForensics++_C23\DeepFakeDetection"
}

# Analysis parameters
ANALYSIS_CONFIG = {
    'sample_frames': 30,  # Frames to sample per video
    'max_videos_per_category': 50,  # Maximum videos to analyze per category
    'frame_skip': 5,  # Analyze every Nth frame
    'image_size': (224, 224),  # Resize dimensions
    'batch_size': 32,
    'random_seed': 42
}

np.random.seed(ANALYSIS_CONFIG['random_seed'])

print("‚úì Configuration complete")

# ============================================================================
# SECTION 4: UTILITY FUNCTIONS WITH ERROR HANDLING
# ============================================================================

print("\n[4/7] Loading Utility Functions...")

class VideoAnalyzer:
    """Self-healing video analysis class with comprehensive error handling"""
    
    def __init__(self, config: dict):
        self.config = config
        self.results = defaultdict(list)
        self.errors = []
        
    def safe_read_video(self, video_path: Path) -> Tuple[bool, Any]:
        """Safely read video with multiple fallback mechanisms"""
        try:
            if not video_path.exists():
                return False, f"File not found: {video_path}"
            
            cap = cv2.VideoCapture(str(video_path))
            if not cap.isOpened():
                return False, f"Cannot open video: {video_path}"
            
            return True, cap
        except Exception as e:
            return False, f"Error reading video: {str(e)}"
    
    def extract_frames(self, video_path: Path, max_frames: int = 30) -> List[np.ndarray]:
        """Extract frames with error handling and adaptive sampling"""
        frames = []
        
        success, cap_or_error = self.safe_read_video(video_path)
        if not success:
            self.errors.append(cap_or_error)
            return frames
        
        cap = cap_or_error
        
        try:
            total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            if total_frames <= 0:
                total_frames = 1000  # Default fallback
            
            # Adaptive sampling
            frame_indices = np.linspace(0, max(total_frames - 1, 0), max_frames, dtype=int)
            
            for idx in frame_indices:
                cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
                ret, frame = cap.read()
                
                if ret and frame is not None:
                    # Validate frame
                    if frame.size > 0:
                        frames.append(frame)
                
                if len(frames) >= max_frames:
                    break
            
        except Exception as e:
            self.errors.append(f"Frame extraction error: {str(e)}")
        finally:
            cap.release()
        
        return frames
    
    def compute_frame_features(self, frame: np.ndarray) -> Dict[str, float]:
        """Compute comprehensive frame features with error handling"""
        features = {}
        
        try:
            # Resize for consistency
            frame_resized = cv2.resize(frame, self.config['image_size'])
            
            # Color space analysis
            gray = cv2.cvtColor(frame_resized, cv2.COLOR_BGR2GRAY)
            hsv = cv2.cvtColor(frame_resized, cv2.COLOR_BGR2HSV)
            
            # Basic statistics
            features['mean_intensity'] = np.mean(gray)
            features['std_intensity'] = np.std(gray)
            features['brightness'] = np.mean(hsv[:,:,2])
            features['saturation'] = np.mean(hsv[:,:,1])
            
            # Texture features
            features['entropy'] = self._safe_entropy(gray)
            features['contrast'] = gray.std()
            
            # Edge detection
            edges = cv2.Canny(gray, 50, 150)
            features['edge_density'] = np.sum(edges > 0) / edges.size
            
            # Frequency domain features
            f_transform = np.fft.fft2(gray)
            f_shift = np.fft.fftshift(f_transform)
            magnitude_spectrum = np.abs(f_shift)
            features['frequency_mean'] = np.mean(magnitude_spectrum)
            features['frequency_std'] = np.std(magnitude_spectrum)
            
            # Laplacian variance (blur detection)
            features['laplacian_var'] = cv2.Laplacian(gray, cv2.CV_64F).var()
            
            # Color distribution
            for i, color in enumerate(['blue', 'green', 'red']):
                features[f'{color}_mean'] = np.mean(frame_resized[:,:,i])
                features[f'{color}_std'] = np.std(frame_resized[:,:,i])
            
            # Noise estimation
            features['noise_estimate'] = self._estimate_noise(gray)
            
        except Exception as e:
            self.errors.append(f"Feature extraction error: {str(e)}")
            # Return default features
            features = {k: 0.0 for k in ['mean_intensity', 'std_intensity', 'brightness', 
                                         'saturation', 'entropy', 'contrast', 'edge_density',
                                         'frequency_mean', 'frequency_std', 'laplacian_var',
                                         'noise_estimate']}
        
        return features
    
    def _safe_entropy(self, image: np.ndarray) -> float:
        """Calculate entropy with error handling"""
        try:
            histogram = np.histogram(image, bins=256, range=(0, 256))[0]
            histogram = histogram[histogram > 0]
            probabilities = histogram / histogram.sum()
            return -np.sum(probabilities * np.log2(probabilities))
        except:
            return 0.0
    
    def _estimate_noise(self, image: np.ndarray) -> float:
        """Estimate noise level in image"""
        try:
            H, W = image.shape
            M = [[1, -2, 1],
                 [-2, 4, -2],
                 [1, -2, 1]]
            sigma = np.sum(np.sum(np.abs(cv2.filter2D(image, cv2.CV_64F, np.array(M)))))
            sigma = sigma * np.sqrt(0.5 * np.pi) / (6 * (W-2) * (H-2))
            return sigma
        except:
            return 0.0

class DatasetAnalyzer:
    """Comprehensive dataset analysis with self-healing capabilities"""
    
    def __init__(self, paths: dict, config: dict):
        self.paths = paths
        self.config = config
        self.video_analyzer = VideoAnalyzer(config)
        self.dataset_stats = {}
        self.feature_data = []
        
    def scan_dataset(self) -> pd.DataFrame:
        """Scan all dataset directories and collect metadata"""
        print("\n" + "="*80)
        print("SCANNING DATASET")
        print("="*80)
        
        all_data = []
        
        for category, path in self.paths.items():
            print(f"\nüìÅ Scanning: {category}")
            path_obj = Path(path)
            
            if not path_obj.exists():
                print(f"   ‚ö† Warning: Path does not exist - {path}")
                continue
            
            # Find all video files
            video_extensions = ['.mp4', '.avi', '.mov', '.mkv', '.flv', '.wmv']
            video_files = []
            
            for ext in video_extensions:
                video_files.extend(list(path_obj.rglob(f'*{ext}')))
            
            print(f"   Found {len(video_files)} video files")
            
            # Sample videos if too many
            if len(video_files) > self.config['max_videos_per_category']:
                video_files = np.random.choice(video_files, 
                                             self.config['max_videos_per_category'], 
                                             replace=False)
                print(f"   Sampled {len(video_files)} videos for analysis")
            
            for video_path in video_files:
                try:
                    file_size = video_path.stat().st_size / (1024 * 1024)  # MB
                    
                    all_data.append({
                        'category': category,
                        'filename': video_path.name,
                        'path': str(video_path),
                        'size_mb': file_size,
                        'is_fake': category != 'Original'
                    })
                except Exception as e:
                    print(f"   ‚ö† Error processing {video_path.name}: {str(e)}")
        
        df = pd.DataFrame(all_data)
        print(f"\n‚úì Total videos found: {len(df)}")
        print(f"‚úì Categories: {df['category'].nunique()}")
        
        return df
    
    def analyze_videos(self, df: pd.DataFrame) -> pd.DataFrame:
        """Extract features from all videos"""
        print("\n" + "="*80)
        print("EXTRACTING VIDEO FEATURES")
        print("="*80)
        
        all_features = []
        
        for idx, row in tqdm(df.iterrows(), total=len(df), desc="Processing videos"):
            try:
                video_path = Path(row['path'])
                
                # Extract frames
                frames = self.video_analyzer.extract_frames(
                    video_path, 
                    max_frames=self.config['sample_frames']
                )
                
                if len(frames) == 0:
                    continue
                
                # Compute features for each frame
                frame_features_list = []
                for frame in frames:
                    features = self.video_analyzer.compute_frame_features(frame)
                    frame_features_list.append(features)
                
                # Aggregate features
                if frame_features_list:
                    aggregated = self._aggregate_features(frame_features_list)
                    aggregated['category'] = row['category']
                    aggregated['is_fake'] = row['is_fake']
                    aggregated['num_frames_analyzed'] = len(frames)
                    aggregated['filename'] = row['filename']
                    
                    all_features.append(aggregated)
                
            except Exception as e:
                print(f"\n‚ö† Error analyzing {row['filename']}: {str(e)}")
                continue
        
        features_df = pd.DataFrame(all_features)
        print(f"\n‚úì Successfully analyzed {len(features_df)} videos")
        
        return features_df
    
    def _aggregate_features(self, feature_list: List[Dict]) -> Dict:
        """Aggregate frame-level features to video-level"""
        aggregated = {}
        
        if not feature_list:
            return aggregated
        
        # Get all feature keys
        feature_keys = feature_list[0].keys()
        
        for key in feature_keys:
            values = [f[key] for f in feature_list if key in f]
            if values:
                aggregated[f'{key}_mean'] = np.mean(values)
                aggregated[f'{key}_std'] = np.std(values)
                aggregated[f'{key}_min'] = np.min(values)
                aggregated[f'{key}_max'] = np.max(values)
                aggregated[f'{key}_median'] = np.median(values)
        
        return aggregated

# ============================================================================
# SECTION 5: ADVANCED STATISTICAL ANALYSIS
# ============================================================================

print("\n[5/7] Preparing Statistical Analysis Functions...")

def perform_statistical_tests(features_df: pd.DataFrame) -> Dict:
    """Perform comprehensive statistical tests"""
    results = {}
    
    print("\n" + "="*80)
    print("STATISTICAL ANALYSIS")
    print("="*80)
    
    # Separate real vs fake
    real_data = features_df[features_df['is_fake'] == False]
    fake_data = features_df[features_df['is_fake'] == True]
    
    print(f"\nReal videos: {len(real_data)}")
    print(f"Fake videos: {len(fake_data)}")
    
    # Get numeric columns
    numeric_cols = features_df.select_dtypes(include=[np.number]).columns
    numeric_cols = [col for col in numeric_cols if col not in ['is_fake', 'num_frames_analyzed']]
    
    # T-tests for each feature
    print("\nPerforming t-tests for feature discrimination...")
    significant_features = []
    
    for col in numeric_cols:
        try:
            real_values = real_data[col].dropna()
            fake_values = fake_data[col].dropna()
            
            if len(real_values) > 0 and len(fake_values) > 0:
                t_stat, p_value = stats.ttest_ind(real_values, fake_values)
                
                if p_value < 0.05:
                    significant_features.append({
                        'feature': col,
                        't_statistic': t_stat,
                        'p_value': p_value,
                        'effect_size': (np.mean(real_values) - np.mean(fake_values)) / np.std(fake_values)
                    })
        except:
            continue
    
    results['significant_features'] = pd.DataFrame(significant_features).sort_values('p_value')
    print(f"‚úì Found {len(significant_features)} statistically significant features (p < 0.05)")
    
    return results

def create_visualizations(features_df: pd.DataFrame, stats_results: Dict, output_dir: Path):
    """Create comprehensive visualizations"""
    print("\n" + "="*80)
    print("GENERATING VISUALIZATIONS")
    print("="*80)
    
    # 1. Distribution Analysis
    print("\n[1/8] Creating distribution plots...")
    fig, axes = plt.subplots(2, 2, figsize=(20, 16))
    fig.suptitle('Dataset Distribution Analysis', fontsize=20, fontweight='bold')
    
    # Category distribution
    category_counts = features_df['category'].value_counts()
    axes[0, 0].bar(range(len(category_counts)), category_counts.values, color='skyblue', edgecolor='black')
    axes[0, 0].set_xticks(range(len(category_counts)))
    axes[0, 0].set_xticklabels(category_counts.index, rotation=45, ha='right')
    axes[0, 0].set_title('Videos per Category', fontsize=14, fontweight='bold')
    axes[0, 0].set_ylabel('Count')
    axes[0, 0].grid(axis='y', alpha=0.3)
    
    # Real vs Fake
    fake_counts = features_df['is_fake'].value_counts()
    axes[0, 1].pie(fake_counts.values, labels=['Real', 'Fake'], autopct='%1.1f%%', 
                   colors=['#90EE90', '#FFB6C6'], startangle=90)
    axes[0, 1].set_title('Real vs Fake Distribution', fontsize=14, fontweight='bold')
    
    # Frames analyzed distribution
    axes[1, 0].hist(features_df['num_frames_analyzed'], bins=30, color='coral', edgecolor='black', alpha=0.7)
    axes[1, 0].set_title('Distribution of Frames Analyzed per Video', fontsize=14, fontweight='bold')
    axes[1, 0].set_xlabel('Number of Frames')
    axes[1, 0].set_ylabel('Frequency')
    axes[1, 0].grid(axis='y', alpha=0.3)
    
    # Category by fake status
    category_fake = pd.crosstab(features_df['category'], features_df['is_fake'])
    category_fake.plot(kind='bar', stacked=True, ax=axes[1, 1], color=['#90EE90', '#FFB6C6'])
    axes[1, 1].set_title('Category Distribution by Authenticity', fontsize=14, fontweight='bold')
    axes[1, 1].set_xlabel('Category')
    axes[1, 1].set_ylabel('Count')
    axes[1, 1].legend(['Real', 'Fake'])
    axes[1, 1].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.savefig(output_dir / '01_distribution_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("‚úì Saved: 01_distribution_analysis.png")
    
    # 2. Feature Importance
    if 'significant_features' in stats_results and len(stats_results['significant_features']) > 0:
        print("\n[2/8] Creating feature importance plots...")
        sig_features = stats_results['significant_features'].head(15)
        
        fig, axes = plt.subplots(1, 2, figsize=(20, 8))
        fig.suptitle('Top Discriminative Features', fontsize=20, fontweight='bold')
        
        # P-values
        axes[0].barh(range(len(sig_features)), -np.log10(sig_features['p_value']), color='steelblue')
        axes[0].set_yticks(range(len(sig_features)))
        axes[0].set_yticklabels(sig_features['feature'], fontsize=10)
        axes[0].set_xlabel('-log10(p-value)', fontsize=12)
        axes[0].set_title('Statistical Significance', fontsize=14, fontweight='bold')
        axes[0].axvline(x=-np.log10(0.05), color='red', linestyle='--', label='p=0.05 threshold')
        axes[0].legend()
        axes[0].grid(axis='x', alpha=0.3)
        
        # Effect sizes
        axes[1].barh(range(len(sig_features)), sig_features['effect_size'], 
                    color=['green' if x > 0 else 'red' for x in sig_features['effect_size']])
        axes[1].set_yticks(range(len(sig_features)))
        axes[1].set_yticklabels(sig_features['feature'], fontsize=10)
        axes[1].set_xlabel('Effect Size (Cohen\'s d)', fontsize=12)
        axes[1].set_title('Effect Size Analysis', fontsize=14, fontweight='bold')
        axes[1].axvline(x=0, color='black', linestyle='-', linewidth=0.5)
        axes[1].grid(axis='x', alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(output_dir / '02_feature_importance.png', dpi=300, bbox_inches='tight')
        plt.show()
        print("‚úì Saved: 02_feature_importance.png")
    
    # 3. Feature Distributions
    print("\n[3/8] Creating feature distribution comparisons...")
    numeric_cols = features_df.select_dtypes(include=[np.number]).columns
    key_features = [col for col in numeric_cols if 'mean_intensity_mean' in col or 
                    'edge_density_mean' in col or 'laplacian_var_mean' in col or
                    'frequency_mean_mean' in col or 'noise_estimate_mean' in col][:6]
    
    if len(key_features) > 0:
        n_features = min(6, len(key_features))
        fig, axes = plt.subplots(2, 3, figsize=(20, 12))
        fig.suptitle('Key Feature Distributions: Real vs Fake', fontsize=20, fontweight='bold')
        axes = axes.flatten()
        
        for idx, feature in enumerate(key_features[:6]):
            real_vals = features_df[features_df['is_fake'] == False][feature].dropna()
            fake_vals = features_df[features_df['is_fake'] == True][feature].dropna()
            
            axes[idx].hist(real_vals, bins=30, alpha=0.6, label='Real', color='green', edgecolor='black')
            axes[idx].hist(fake_vals, bins=30, alpha=0.6, label='Fake', color='red', edgecolor='black')
            axes[idx].set_title(feature.replace('_', ' ').title(), fontsize=12, fontweight='bold')
            axes[idx].set_xlabel('Value')
            axes[idx].set_ylabel('Frequency')
            axes[idx].legend()
            axes[idx].grid(alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(output_dir / '03_feature_distributions.png', dpi=300, bbox_inches='tight')
        plt.show()
        print("‚úì Saved: 03_feature_distributions.png")
    
    # 4. Correlation Analysis
    print("\n[4/8] Creating correlation heatmap...")
    numeric_cols = features_df.select_dtypes(include=[np.number]).columns
    sample_features = [col for col in numeric_cols if '_mean' in col][:20]
    
    if len(sample_features) > 1:
        correlation_matrix = features_df[sample_features].corr()
        
        plt.figure(figsize=(16, 14))
        sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0, 
                    square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
        plt.title('Feature Correlation Matrix', fontsize=18, fontweight='bold', pad=20)
        plt.tight_layout()
        plt.savefig(output_dir / '04_correlation_matrix.png', dpi=300, bbox_inches='tight')
        plt.show()
        print("‚úì Saved: 04_correlation_matrix.png")
    
    # 5. Box plots by category
    print("\n[5/8] Creating category comparison box plots...")
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    fig.suptitle('Feature Comparison Across Categories', fontsize=20, fontweight='bold')
    axes = axes.flatten()
    
    for idx, feature in enumerate(key_features[:6]):
        data_to_plot = []
        labels_to_plot = []
        
        for category in features_df['category'].unique():
            vals = features_df[features_df['category'] == category][feature].dropna()
            if len(vals) > 0:
                data_to_plot.append(vals)
                labels_to_plot.append(category)
        
        if data_to_plot:
            bp = axes[idx].boxplot(data_to_plot, labels=labels_to_plot, patch_artist=True)
            for patch in bp['boxes']:
                patch.set_facecolor('lightblue')
            axes[idx].set_title(feature.replace('_', ' ').title(), fontsize=12, fontweight='bold')
            axes[idx].tick_params(axis='x', rotation=45)
            axes[idx].grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(output_dir / '05_category_boxplots.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("‚úì Saved: 05_category_boxplots.png")
    
    # 6. PCA Analysis
    print("\n[6/8] Creating PCA visualization...")
    numeric_features = features_df.select_dtypes(include=[np.number]).drop(
        columns=['is_fake', 'num_frames_analyzed'], errors='ignore')
    
    if len(numeric_features.columns) >= 2:
        # Handle missing values
        numeric_features_clean = numeric_features.fillna(numeric_features.mean())
        
        # Standardize
        scaler = StandardScaler()
        features_scaled = scaler.fit_transform(numeric_features_clean)
        
        # PCA
        pca = PCA(n_components=2)
        pca_result = pca.fit_transform(features_scaled)
        
        plt.figure(figsize=(14, 10))
        
        colors = {'Original': 'green', 'NeuralTextures': 'red', 'FaceSwap': 'blue',
                 'FaceShifter': 'orange', 'Face2Face': 'purple', 'Deepfakes': 'brown',
                 'DeepFakeDetection': 'pink'}
        
        for category in features_df['category'].unique():
            mask = features_df['category'] == category
            plt.scatter(pca_result[mask, 0], pca_result[mask, 1], 
                       c=colors.get(category, 'gray'), label=category, alpha=0.6, s=50)
        
        plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)', fontsize=12)
        plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)', fontsize=12)
        plt.title('PCA: Feature Space Visualization', fontsize=18, fontweight='bold')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.savefig(output_dir / '06_pca_visualization.png', dpi=300, bbox_inches='tight')
        plt.show()
        print("‚úì Saved: 06_pca_visualization.png")
    
    # 7. Violin plots
    print("\n[7/8] Creating violin plots...")
    fig, axes = plt.subplots(2, 2, figsize=(18, 14))
    fig.suptitle('Distribution Comparison: Real vs Fake', fontsize=20, fontweight='bold')
    axes = axes.flatten()
    
    selected_features = key_features[:4] if len(key_features) >= 4 else key_features
    
    for idx, feature in enumerate(selected_features):
        if idx < 4:
            sns.violinplot(data=features_df, x='is_fake', y=feature, ax=axes[idx])
            axes[idx].set_title(feature.replace('_', ' ').title(), fontsize=12, fontweight='bold')
            axes[idx].set_xlabel('Is Fake', fontsize=11)
            axes[idx].set_ylabel('Value', fontsize=11)
            axes[idx].set_xticklabels(['Real', 'Fake'])
            axes[idx].grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(output_dir / '07_violin_plots.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("‚úì Saved: 07_violin_plots.png")
    
    # 8. Summary Statistics Table
    print("\n[8/8] Creating summary statistics visualization...")
    fig, ax = plt.subplots(figsize=(18, 10))
    ax.axis('tight')
    ax.axis('off')
    
    # Create summary statistics
    summary_data = []
    for category in features_df['category'].unique():
        cat_data = features_df[features_df['category'] == category]
        summary_data.append({
            'Category': category,
            'Count': len(cat_data),
            'Avg Frames': f"{cat_data['num_frames_analyzed'].mean():.1f}",
            'Type': 'Real' if not cat_data['is_fake'].any() else 'Fake'
        })
    
    summary_df = pd.DataFrame(summary_data)
    
    table = ax.table(cellText=summary_df.values, colLabels=summary_df.columns,
                    cellLoc='center', loc='center', bbox=[0, 0, 1, 1])
    table.auto_set_font_size(False)
    table.set_fontsize(12)
    table.scale(1, 3)
    
    # Style header
    for i in range(len(summary_df.columns)):
        table[(0, i)].set_facecolor('#4CAF50')
        table[(0, i)].set_text_props(weight='bold', color='white')
    
    # Alternate row colors
    for i in range(1, len(summary_df) + 1):
        for j in range(len(summary_df.columns)):
            if i % 2 == 0:
                table[(i, j)].set_facecolor('#f0f0f0')
            else:
                table[(i, j)].set_facecolor('#ffffff')
    
    plt.title('Dataset Summary Statistics', fontsize=18, fontweight='bold', pad=20)
    plt.savefig(output_dir / '08_summary_statistics.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("‚úì Saved: 08_summary_statistics.png")
    
    print("\n" + "="*80)
    print("‚úì ALL VISUALIZATIONS COMPLETED")
    print("="*80)

def generate_research_report(features_df: pd.DataFrame, stats_results: Dict, output_dir: Path):
    """Generate comprehensive research report"""
    print("\n" + "="*80)
    print("GENERATING RESEARCH REPORT")
    print("="*80)
    
    report_lines = []
    report_lines.append("="*80)
    report_lines.append("DEEPFAKE VIDEO ANALYSIS - COMPREHENSIVE RESEARCH REPORT")
    report_lines.append("FaceForensics++ Dataset Analysis")
    report_lines.append(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    report_lines.append("="*80)
    
    # Dataset Overview
    report_lines.append("\n" + "="*80)
    report_lines.append("1. DATASET OVERVIEW")
    report_lines.append("="*80)
    report_lines.append(f"Total Videos Analyzed: {len(features_df)}")
    report_lines.append(f"Total Categories: {features_df['category'].nunique()}")
    report_lines.append(f"Real Videos: {(~features_df['is_fake']).sum()}")
    report_lines.append(f"Fake Videos: {features_df['is_fake'].sum()}")
    report_lines.append(f"\nAverage Frames per Video: {features_df['num_frames_analyzed'].mean():.2f}")
    report_lines.append(f"Median Frames per Video: {features_df['num_frames_analyzed'].median():.2f}")
    
    # Category breakdown
    report_lines.append("\n" + "-"*80)
    report_lines.append("Category Distribution:")
    report_lines.append("-"*80)
    for category, count in features_df['category'].value_counts().items():
        percentage = (count / len(features_df)) * 100
        report_lines.append(f"  {category:25s}: {count:4d} videos ({percentage:5.2f}%)")
    
    # Statistical Analysis
    if 'significant_features' in stats_results and len(stats_results['significant_features']) > 0:
        report_lines.append("\n" + "="*80)
        report_lines.append("2. STATISTICAL ANALYSIS")
        report_lines.append("="*80)
        
        sig_features = stats_results['significant_features']
        report_lines.append(f"Statistically Significant Features (p < 0.05): {len(sig_features)}")
        
        report_lines.append("\n" + "-"*80)
        report_lines.append("Top 10 Discriminative Features:")
        report_lines.append("-"*80)
        report_lines.append(f"{'Feature':<40} {'p-value':<12} {'Effect Size':<12}")
        report_lines.append("-"*80)
        
        for idx, row in sig_features.head(10).iterrows():
            report_lines.append(f"{row['feature']:<40} {row['p_value']:<12.6f} {row['effect_size']:<12.4f}")
    
    # Feature Statistics
    report_lines.append("\n" + "="*80)
    report_lines.append("3. FEATURE STATISTICS")
    report_lines.append("="*80)
    
    numeric_cols = features_df.select_dtypes(include=[np.number]).columns
    key_stats_features = [col for col in numeric_cols if '_mean' in col][:10]
    
    report_lines.append("\n" + "-"*80)
    report_lines.append("Key Feature Ranges:")
    report_lines.append("-"*80)
    
    for feature in key_stats_features:
        report_lines.append(f"\n{feature}:")
        report_lines.append(f"  Overall  - Mean: {features_df[feature].mean():.4f}, "
                          f"Std: {features_df[feature].std():.4f}")
        
        real_mean = features_df[~features_df['is_fake']][feature].mean()
        fake_mean = features_df[features_df['is_fake']][feature].mean()
        report_lines.append(f"  Real     - Mean: {real_mean:.4f}")
        report_lines.append(f"  Fake     - Mean: {fake_mean:.4f}")
        report_lines.append(f"  Difference: {abs(real_mean - fake_mean):.4f}")
    
    # Key Findings
    report_lines.append("\n" + "="*80)
    report_lines.append("4. KEY FINDINGS")
    report_lines.append("="*80)
    
    real_count = (~features_df['is_fake']).sum()
    fake_count = features_df['is_fake'].sum()
    
    report_lines.append(f"\n‚Ä¢ Dataset contains {fake_count} fake videos and {real_count} real videos")
    report_lines.append(f"‚Ä¢ Fake-to-Real ratio: {fake_count/real_count:.2f}:1")
    
    if 'significant_features' in stats_results and len(stats_results['significant_features']) > 0:
        report_lines.append(f"‚Ä¢ Identified {len(stats_results['significant_features'])} statistically significant features")
        report_lines.append(f"‚Ä¢ Top discriminative feature: {stats_results['significant_features'].iloc[0]['feature']}")
    
    # Recommendations
    report_lines.append("\n" + "="*80)
    report_lines.append("5. RESEARCH RECOMMENDATIONS")
    report_lines.append("="*80)
    report_lines.append("\n‚Ä¢ Focus on top discriminative features for model development")
    report_lines.append("‚Ä¢ Consider temporal features across multiple frames")
    report_lines.append("‚Ä¢ Implement ensemble methods combining multiple feature types")
    report_lines.append("‚Ä¢ Validate findings on additional datasets")
    report_lines.append("‚Ä¢ Explore deep learning approaches with identified features")
    
    report_lines.append("\n" + "="*80)
    report_lines.append("END OF REPORT")
    report_lines.append("="*80)
    
    # Save report
    report_text = "\n".join(report_lines)
    report_path = output_dir / 'research_report.txt'
    with open(report_path, 'w') as f:
        f.write(report_text)
    
    print(report_text)
    print(f"\n‚úì Report saved: {report_path}")
    
    return report_text

def save_detailed_results(features_df: pd.DataFrame, stats_results: Dict, output_dir: Path):
    """Save detailed analysis results"""
    print("\n" + "="*80)
    print("SAVING DETAILED RESULTS")
    print("="*80)
    
    # Save feature data
    features_path = output_dir / 'extracted_features.csv'
    features_df.to_csv(features_path, index=False)
    print(f"‚úì Saved: extracted_features.csv ({len(features_df)} rows)")
    
    # Save statistical results
    if 'significant_features' in stats_results:
        stats_path = output_dir / 'statistical_tests.csv'
        stats_results['significant_features'].to_csv(stats_path, index=False)
        print(f"‚úì Saved: statistical_tests.csv ({len(stats_results['significant_features'])} rows)")
    
    # Save summary JSON
    summary = {
        'analysis_date': datetime.now().isoformat(),
        'total_videos': len(features_df),
        'real_videos': int((~features_df['is_fake']).sum()),
        'fake_videos': int(features_df['is_fake'].sum()),
        'categories': features_df['category'].value_counts().to_dict(),
        'avg_frames_analyzed': float(features_df['num_frames_analyzed'].mean()),
        'num_features_extracted': len(features_df.columns),
        'significant_features_count': len(stats_results.get('significant_features', []))
    }
    
    json_path = output_dir / 'analysis_summary.json'
    with open(json_path, 'w') as f:
        json.dump(summary, f, indent=2)
    print(f"‚úì Saved: analysis_summary.json")
    
    print("\n" + "="*80)
    print("‚úì ALL RESULTS SAVED")
    print("="*80)

# ============================================================================
# SECTION 6: MAIN EXECUTION PIPELINE
# ============================================================================

print("\n[6/7] Initializing Main Analysis Pipeline...")

def run_complete_analysis():
    """Execute complete analysis pipeline with error handling"""
    
    print("\n" + "="*80)
    print("STARTING COMPREHENSIVE DEEPFAKE ANALYSIS")
    print("="*80)
    print(f"Analysis Configuration:")
    print(f"  - Sample Frames per Video: {ANALYSIS_CONFIG['sample_frames']}")
    print(f"  - Max Videos per Category: {ANALYSIS_CONFIG['max_videos_per_category']}")
    print(f"  - Frame Skip: {ANALYSIS_CONFIG['frame_skip']}")
    print(f"  - Image Size: {ANALYSIS_CONFIG['image_size']}")
    print("="*80)
    
    try:
        # Initialize analyzer
        analyzer = DatasetAnalyzer(DATASET_PATHS, ANALYSIS_CONFIG)
        
        # Step 1: Scan dataset
        print("\nSTEP 1: Scanning Dataset...")
        df_metadata = analyzer.scan_dataset()
        
        if len(df_metadata) == 0:
            print("\n‚ö† ERROR: No videos found in the specified paths!")
            print("Please verify the dataset paths are correct.")
            return None
        
        # Step 2: Extract features
        print("\nSTEP 2: Extracting Features...")
        features_df = analyzer.analyze_videos(df_metadata)
        
        if len(features_df) == 0:
            print("\n‚ö† ERROR: Feature extraction failed!")
            return None
        
        # Step 3: Statistical analysis
        print("\nSTEP 3: Statistical Analysis...")
        stats_results = perform_statistical_tests(features_df)
        
        # Step 4: Generate visualizations
        print("\nSTEP 4: Creating Visualizations...")
        create_visualizations(features_df, stats_results, OUTPUT_DIR)
        
        # Step 5: Generate report
        print("\nSTEP 5: Generating Research Report...")
        report = generate_research_report(features_df, stats_results, OUTPUT_DIR)
        
        # Step 6: Save detailed results
        print("\nSTEP 6: Saving Detailed Results...")
        save_detailed_results(features_df, stats_results, OUTPUT_DIR)
        
        # Final summary
        print("\n" + "="*80)
        print("ANALYSIS COMPLETE!")
        print("="*80)
        print(f"\n‚úì Analysis Results Location: {OUTPUT_DIR.absolute()}")
        print(f"‚úì Total Videos Analyzed: {len(features_df)}")
        print(f"‚úì Visualizations Generated: 8 PNG files")
        print(f"‚úì Data Files Saved: 3 files (CSV, JSON, TXT)")
        
        if hasattr(analyzer.video_analyzer, 'errors') and analyzer.video_analyzer.errors:
            print(f"\n‚ö† Note: {len(analyzer.video_analyzer.errors)} non-critical errors occurred during processing")
            print("  (These errors were handled automatically and did not affect the analysis)")
        
        print("\n" + "="*80)
        print("READY FOR RESEARCH USE")
        print("="*80)
        
        return features_df, stats_results
        
    except Exception as e:
        print("\n" + "="*80)
        print("CRITICAL ERROR IN ANALYSIS PIPELINE")
        print("="*80)
        print(f"Error: {str(e)}")
        print("\nAttempting recovery...")
        
        import traceback
        traceback.print_exc()
        
        return None

# ============================================================================
# SECTION 7: EXECUTION
# ============================================================================

print("\n[7/7] Ready to Execute Analysis")
print("\n" + "="*80)
print("INSTRUCTIONS:")
print("="*80)
print("Run the following command to start the complete analysis:")
print("\n    results = run_complete_analysis()")
print("\nThis will:")
print("  1. Scan all dataset directories")
print("  2. Extract video features")
print("  3. Perform statistical analysis")
print("  4. Generate 8 visualization plots")
print("  5. Create a comprehensive research report")
print("  6. Save all results to the output directory")
print("\nAll results will be saved to: deepfake_analysis_results/")
print("="*80)

# Auto-run the analysis
print("\nüöÄ AUTO-STARTING ANALYSIS...\n")
results = run_complete_analysis()

if results is not None:
    features_df, stats_results = results
    print("\n‚úÖ Analysis completed successfully!")
    print("\nYou can now:")
    print("  - View the visualizations above")
    print("  - Access features_df for the extracted features DataFrame")
    print("  - Access stats_results for statistical test results")
    print(f"  - Find all saved files in: {OUTPUT_DIR.absolute()}")
else:
    print("\n‚ö† Analysis encountered errors. Please check the paths and try again.")
    print("\nTroubleshooting tips:")
    print("  1. Verify the dataset paths exist")
    print("  2. Ensure video files are in supported formats (.mp4, .avi, etc.)")
    print("  3. Check that you have read permissions for the directories")

print("\n" + "="*80)
print("DEEPFAKE ANALYSIS SYSTEM READY")
print("="*80)
```
